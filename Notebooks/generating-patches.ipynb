```python
import cv2
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm, trange
from joblib import Parallel, delayed
from patch_gen import PatchGenerator

%matplotlib inline
```

```python
path = "/kaggle/input/cbisddsm/CBIS_DDSM/"
```

```python
patch_gen = PatchGenerator(path, patch_size=(256,256), max_patches=50, min_ratio=0.10, max_black_space=0.5, max_trials=100)
```

```python
i = np.random.randint(0, len(patch_gen.imgs_path))
example = patch_gen.imgs_path[i]
```

```python
bg_patch = patch_gen._create_background_patch(example)
mass_patch, _, _ = patch_gen._create_mass_patch(example, 0.10)
```

```python
_, axs = plt.subplots(1,2)

axs[0].imshow(bg_patch, cmap="bone")
axs[0].set_title("Background patch")

axs[1].imshow(mass_patch, cmap="bone")
axs[1].set_title("Mass patch")

plt.tight_layout();
```
